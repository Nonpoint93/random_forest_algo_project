```python
# Your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score

X_train = pd.read_csv('../models/X_train_ts_02_rs_42.csv')
y_train = pd.read_csv('../models/y_train_ts_02_rs_42.csv').values.ravel()
X_test = pd.read_csv('../models/X_test_ts_02_rs_42.csv')
y_test = pd.read_csv('../models/y_test_ts_02_rs_42.csv').values.ravel()


X_train.head()
```

```python
random_forest_model = RandomForestClassifier(n_estimators=10, random_state=42)
random_forest_model.fit(X_train, y_train)

fig, axis = plt.subplots(2, 2, figsize = (15, 15))

tree.plot_tree(random_forest_model.estimators_[0], ax = axis[0, 0], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(random_forest_model.estimators_[1], ax = axis[0, 1], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(random_forest_model.estimators_[2], ax = axis[1, 0], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(random_forest_model.estimators_[3], ax = axis[1, 1], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)
plt.show()
```

```python
y_pred = random_forest_model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)

```
